```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load
file = "avocado (version 1).csv"

# Read Purchasing File and store into Pandas data frame
avocado = pd.read_csv(file)
avocado.head()
```

```python
avocado.drop('0', axis=1, inplace=True)

avocado.head()
```

```python
avocado.drop(avocado[avocado['region'] == 'Mid South'].index, inplace = True)
avocado.head()
```

```python
avocado.drop(avocado[avocado['region'] == 'North East'].index, inplace = True)
avocado.head()
```

```python
avocado.drop(avocado[avocado['region'] == 'South Central'].index, inplace = True)
avocado.head()
```

```python
avocado.drop(avocado[avocado['region'] == 'South East'].index, inplace = True)
avocado.head()
```

```python
avocado.drop(avocado[avocado['region'] == 'Total US'].index, inplace = True)
avocado.head()
```

```python
avocado.drop(avocado[avocado['region'] == 'West'].index, inplace = True)
```

```python
avocado=avocado.rename(columns = {'region':'city'})
avocado.head()
```

```python
avocado = avocado.rename(columns={'Date':'date', 'AveragePrice':'average_price', 'Total Volume':'total_volume', '4046':'PLU 4046', '4225':'PLU 4225', '4770':'PLU 4770'}) 
avocado.head()
```

```python
avocado = avocado.rename(columns={'PLU 4046':'plu_4046_volume', 'PLU 4225':'plu_4225_volume', 'PLU 4770':'plu_4770_volume'}) 
avocado.head()
```

```python
avocado.drop(avocado[avocado['city'] == 'California'].index, inplace = True)
```

```python
avocado.drop(avocado[avocado['city'] == 'Plains'].index, inplace = True)
avocado.drop(avocado[avocado['city'] == 'South Carolina'].index, inplace = True)
avocado.drop(avocado[avocado['city'] == 'West'].index, inplace = True)
```
