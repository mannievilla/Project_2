```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)
file_to_load = "population.csv"

# Read Purchasing File and store into Pandas data frame
population = pd.read_csv(file_to_load)
population.head()
```

```python
population=population.rename(columns = {'geographic area':'city_state'})
```

```python
# pd.Series(['city']).str.replace(' ', regex=True)
```

```python
population
```

```python
population['city_state'] = population.city_state.replace({'city' : ' '}, regex=True)
```

```python
population['city_state'] = population.city_state.replace({'town': ' '}, regex=True)
```

```python
population['city_state'] = population.city_state.replace({'village': ' '}, regex=True)
```

```python
population['city_state'] = population.city_state.replace({'borough': ' '}, regex=True)
```

```python
population
```

```python
population.to_csv("population.csv")
```
